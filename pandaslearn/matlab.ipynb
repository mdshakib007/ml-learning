```python
%matplotlib inline
import matplotlib.pyplot as plt # similar: 'from matplotlib import pyplot as plt'
```

```python
plt.plot([1,4,3], [4,5,7]) # first list refer 'x'-axis, and second list refer 'y'-axis.
```

```python
plt.plot([1,2,3], [2,3,5])# first list refer 'x'-axis, and second list refer 'y'-axis.
plt.title('my data')
plt.xlabel('this is x')
plt.ylabel('this is y')
plt.show()
```

```python
from matplotlib import style
style.use('ggplot') # this is style for plot

x1 = [2,4,5]
y1 = [5,3,6]
x2 = [34,54,23]
y2 = [23,34,56]


plt.plot(x1,y1, label='first')
plt.plot(x2,y2, label='second')
plt.legend()
plt.show()
```
